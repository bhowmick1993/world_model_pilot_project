```python
import re
import sys
from pathlib import Path

try:
    from PIL import Image
except ImportError:
    %pip install pillow -q
    from PIL import Image

folder_of_images = Path(r"C:\AB_Personal\linkedin_articles\MDP")
output_gif = folder_of_images / "mdp_animation.gif"

image_extensions = {".png", ".jpg", ".jpeg", ".webp", ".bmp"}


def natural_sort_key(path: Path) -> list:
    """Sort filenames so (2) comes before (10)."""
    return [int(part) if part.isdigit() else part.lower() for part in re.split(r"(\d+)", path.name)]


image_paths = sorted(
    (p for p in folder_of_images.iterdir() if p.is_file() and p.suffix.lower() in image_extensions),
    key=natural_sort_key,
)

print(f"Found {len(image_paths)} images:")
for path in image_paths:
    print(f"  {path.name}")

frames = [Image.open(path).convert("RGBA") for path in image_paths]
base_size = frames[0].size
frames = [
    frame.resize(base_size, Image.Resampling.LANCZOS) if frame.size != base_size else frame
    for frame in frames
]
frames_rgb = [frame.convert("RGB") for frame in frames]

duration_ms = 800  # milliseconds per frame
frames_rgb[0].save(
    output_gif,
    save_all=True,
    append_images=frames_rgb[1:],
    duration=duration_ms,
    loop=0,
)

print(f"\nSaved GIF ({len(frames_rgb)} frames) to:\n  {output_gif}")
```
